```python
#converting string of list to a list before calling the convert function
#import ast
#ast.literal_eval()
```

```python
#function to pre-process the data for genres and keywords

def convert(obj):
    Lst = []
    for i in ast.literal_eval(obj):
        Lst.append(i['name'])
    return Lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#function to pre-process the data for cast ; taking the three lead actors / actresses

def convert_cast(obj):
    Lst = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            Lst.append(i['name'])
            counter=counter+1
        else:
            break
    return Lst
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
#function to pre-process the data for director of the movie

def find_director(obj):
    Lst = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            Lst.append(i['name'])
            break
    return Lst
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.head()
```

```python
#making the overview column a string

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing all the unnecessary spaces in the data

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#creating a column of all the tags

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#creating new data-frame

new_data = movies[['movie_id' , 'title' , 'tags']]
```

```python
new_data
```

```python
#converting the list of tags into string

new_data['tags'] = new_data['tags'].apply(lambda x:" ".join(x))
```

```python
new_data['tags'][0]
```

```python
#converting the entire string into lower_case

new_data['tags'] = new_data['tags'].apply(lambda x:x.lower())
```

```python
new_data['tags'][0]
```

```python
new_data.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#function to remove similar words ; stemming

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_data['tags'] = new_data['tags'].apply(stem)
```

```python
#Performing Text vectorization over the tags

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#finding similarity of vectors ; in terms of cosine distance

similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
#defining the recommender function
#distances variable is similarity of the index of the recieved movie w.r.t all other movies

def recommend(movie):
    movie_index = new_data[new_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('Batman Returns')
```
